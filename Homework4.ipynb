```python
# Creating Bins 
bins = [0,9,14,19,24,29,34,39, 45]
group_names = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
                               
                            

#Creating Total Counts and Percentages

shortend_df = purchase_data.drop_duplicates(['SN'])



age_range_count_df = shortend_df.groupby("Age Range")
age_range_count_df.count()
age_new_df = pd.DataFrame(shortend_df.groupby("Age Range").count())

total_count_df = age_new_df.rename(columns={"Purchase ID": "Total Count"})
del total_count_df['Age']
del total_count_df['Gender']
del total_count_df['Item ID']
del total_count_df['Price']
del total_count_df['SN']
del total_count_df['Item Name']
total_count_df
#len(shortend_df)

percentages = []
my_list = total_count_df["Total Count"].tolist()
my_list

for num in my_list:
    percent = round(((num/len(unique_SNs))*100),2)
    percentages.append(percent)

percentages

total_count_df['Percentage of Players'] = percentages
total_count_df
```

```python
# Purchasing Analysis (Age)
new_purchase_data = purchase_data.set_index('Age Range')
new_purchase_data


age_purchase_total = new_purchase_data.groupby(["Age Range"]).sum()["Price"]
average_purchase_price = new_purchase_data.groupby(["Age Range"]).mean()["Price"]
purchase_counts = new_purchase_data.groupby(["Age Range"]).count()["Price"]
Avg_Total_Purchase_per_Person = age_purchase_total / total_count_df["Total Count"]


#Making the Data Frame
Purchasing_Analysis = pd.DataFrame({
    "Purchase Counts": (purchase_counts),
    "Average Purchase Price": round((average_purchase_price),2),
    "Total Purchase Value": (age_purchase_total),
    "Avg Total Purchase per Person": round((Avg_Total_Purchase_per_Person),2)
})

Purchasing_Analysis
```

```python
# Top Spenders

purchase_count_top_spenders = purchase_data.groupby(['SN']).count()['Price']
Average_purchase_price_top_spenders = purchase_data.groupby(['SN']).mean()['Price']
total_purcahse_value_top_spenders = purchase_data.groupby(['SN']).sum()['Price']

top_spenders_df = pd.DataFrame({
    "Purchase Count": (purchase_count_top_spenders),
    "Average Purchase Price": (Average_purchase_price_top_spenders),
    "Total Purchase Value": (total_purcahse_value_top_spenders)
})

top_spenders_df.sort_values("Total Purchase Value", ascending=False).head(5)

```

```python
# Most Popular Items 

item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

purchase_count_pop_items = item_data.groupby(['Item ID', 'Item Name']).count()['Price']
Total_Purchase_Value = item_data.groupby(['Item ID', 'Item Name']).sum()['Price']
#Item_Price = item_data['Price'].apply(Price)
#Item_Price = purchase_data.iloc[:, 6]




most_popular_df = pd.DataFrame({
    "Purchase Count": (purchase_count_pop_items),
    #"Item Price": Item_Price,
    "Total Purchase Value": (Total_Purchase_Value)
})

most_popular_df.sort_values("Total Purchase Value", ascending=False).head(5)
Merged_df = pd.merge(item_data, most_popular_df, on="Item ID")
Merged_df_clean = Merged_df.drop_duplicates().sort_values("Purchase Count", ascending=False)
Merged_df_clean.head()
```
